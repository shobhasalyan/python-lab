```python
# Function to convert decimal to binary without built-in function
def decimal_to_binary(decimal_num):
    binary_num = ''
    while decimal_num > 0:
        binary_num = str(decimal_num % 2) + binary_num
        decimal_num //= 2
    return binary_num

# Function to convert decimal to octal without built-in function
def decimal_to_octal(decimal_num):
    octal_num = ''
    while decimal_num > 0:
        octal_num = str(decimal_num % 8) + octal_num
        decimal_num //= 8
    return octal_num

# Function to convert decimal to hexadecimal without built-in function
def decimal_to_hexadecimal(decimal_num):
    hexadecimal_num = ''
    while decimal_num > 0:
        remainder = decimal_num % 16
        if remainder < 10:
            hexadecimal_num = str(remainder) + hexadecimal_num
        else:
            hexadecimal_num = chr(remainder + 55) + hexadecimal_num
        decimal_num //= 16
    return hexadecimal_num

# Function to convert binary to decimal without built-in function
def binary_to_decimal(binary_num):
    decimal_num = 0
    power = 0
    for digit in binary_num[::-1]:
        decimal_num += int(digit) * (2 ** power)
        power += 1
    return decimal_num

# Function to convert octal to decimal without built-in function
def octal_to_decimal(octal_num):
    decimal_num = 0
    power = 0
    for digit in octal_num[::-1]:
        decimal_num += int(digit) * (8 ** power)
        power += 1
    return decimal_num

# Function to convert hexadecimal to decimal without built-in function
def hexadecimal_to_decimal(hexadecimal_num):
    decimal_num = 0
    power = 0
    for digit in hexadecimal_num[::-1]:
        if digit.isdigit():
            decimal_num += int(digit) * (16 ** power)
        else:
            decimal_num += (ord(digit.upper()) - 55) * (16 ** power)
        power += 1
    return decimal_num

# Function to convert binary to octal without built-in function
def binary_to_octal(binary_num):
    decimal_num = binary_to_decimal(binary_num)
    return decimal_to_octal(decimal_num)

# Function to convert binary to hexadecimal without built-in function
def binary_to_hexadecimal(binary_num):
    decimal_num = binary_to_decimal(binary_num)
    return decimal_to_hexadecimal(decimal_num)

# Function to convert octal to binary without built-in function
def octal_to_binary(octal_num):
    decimal_num = octal_to_decimal(octal_num)
    return decimal_to_binary(decimal_num)

# Function to convert octal to hexadecimal without built-in function
def octal_to_hexadecimal(octal_num):
    decimal_num = octal_to_decimal(octal_num)
    return decimal_to_hexadecimal(decimal_num)

# Function to convert hexadecimal to binary without built-in function
def hexadecimal_to_binary(hexadecimal_num):
    decimal_num = hexadecimal_to_decimal(hexadecimal_num)
    return decimal_to_binary(decimal_num)

# Function to convert hexadecimal to octal without built-in function
def hexadecimal_to_octal(hexadecimal_num):
    decimal_num = hexadecimal_to_decimal(hexadecimal_num)
    return decimal_to_octal(decimal_num)

# Function to display all conversions
def display_all_conversions(number):
    print("Decimal:", number)
    print("Binary:", decimal_to_binary(number))
    print("Octal:", decimal_to_octal(number))
    print("Hexadecimal:", decimal_to_hexadecimal(number))
    print()

# Main function
def main():
    print("\nMenu:")
    print("DECIMAL")
    print("1. Decimal to Binary")
    print("2. Decimal to Octal")
    print("3. Decimal to Hexadecimal")
    print("4. Binary to Decimal")
    print("5. Octal to Decimal")
    print("6. Hexadecimal to Decimal")
    print("7. Binary to Octal")
    print("8. Binary to Hexadecimal")
    print("9. Octal to Binary")
    print("10. Octal to Hexadecimal")
    print("11. Hexadecimal to Binary")
    print("12. Hexadecimal to Octal")
    print("13. Display all conversions for a number")
    print("14. Exit")    
    while True:
        choice = input("Enter your choice: ")
        
        if choice == '1':
            decimal_num = int(input("Enter the decimal number: "))
            print("Binary:", decimal_to_binary(decimal_num))
            print("Octal:", decimal_to_octal(decimal_num))
            print("Hexadecimal:", decimal_to_hexadecimal(decimal_num))
        elif choice == '2':
            binary_num = input("Enter the binary number: ")
            print("Decimal:", binary_to_decimal(binary_num))
            print("Octal:", binary_to_octal(binary_num))
            print("Hexadecimal:", binary_to_hexadecimal(binary_num))
        elif choice == '3':
            octal_num = input("Enter the octal number: ")
            print("Decimal:", octal_to_decimal(octal_num))
            print("Binary:", octal_to_binary(octal_num))
            print("Hexadecimal:", octal_to_hexadecimal(octal_num))
        elif choice == '4':
            hexadecimal_num = input("Enter the hexadecimal number: ")
            print("Decimal:", hexadecimal_to_decimal(hexadecimal_num))
            print("Binary:", hexadecimal_to_binary(hexadecimal_num))
            print("Octal:", hexadecimal_to_octal(hexadecimal_num))
        elif choice == '5':
            number = int(input("Enter a number: "))
            display_all_conversions(number)
        elif choice == '6':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")
        print("\n")

if __name__ == "__main__":
    main()
```

```python
import math
import numpy

# Function to calculate mean using built-in function
def calculate_mean_builtin(data):
    return sum(data) / len(data)

# Function to calculate variance using built-in function
def calculate_variance_builtin(data):
    mean = calculate_mean_builtin(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    return variance

# Function to calculate standard deviation using built-in function
def calculate_standard_deviation_builtin(data):
    variance = calculate_variance_builtin(data)
    return math.sqrt(variance)

# Function to calculate quartiles using built-in function
def calculate_quartiles_builtin(data):
    Q1 = numpy.percentile(data, 25, interpolation='lower')
    Q3 = numpy.percentile(data, 75, interpolation='higher')
    return Q1, Q3

# Function to calculate IQR using built-in function
def calculate_iqr_builtin(data):
    Q1, Q3 = calculate_quartiles_builtin(data)
    return Q3 - Q1

# Function to take user input for data
def take_user_input():
    data = []
    while True:
        num = input("Enter a number (or 'done' to finish): ")
        if num.lower() == 'done':
            break
        data.append(float(num))
    return data

# Main function
def main():
    data = take_user_input()
    print("Using built-in functions:")
    print("Mean:", calculate_mean_builtin(data))
    print("Variance:", calculate_variance_builtin(data))
    print("Standard Deviation:", calculate_standard_deviation_builtin(data))
    print("Quartiles:", calculate_quartiles_builtin(data))
    print("IQR:", calculate_iqr_builtin(data))

if __name__ == "__main__":
    main()
```

```python
# Function to calculate mean without using built-in function
def calculate_mean(data):
    total = 0
    count = 0
    for num in data:
        total += num
        count += 1
    return total / count

# Function to calculate variance without using built-in function
def calculate_variance(data):
    mean = calculate_mean(data)
    total_diff_squared = 0
    count = 0
    for num in data:
        total_diff_squared += (num - mean) ** 2
        count += 1
    return total_diff_squared / count

# Function to calculate standard deviation without using built-in function
def calculate_standard_deviation(data):
    return calculate_variance(data) ** 0.5

# Function to calculate quartiles without using built-in function
# Function to calculate quartiles without using built-in function
def calculate_quartiles(data):
    data_copy = data[:]
    data_copy.sort()
    n = 0
    for num in data_copy:
        n += 1
    Q1_index = (n + 1) // 4
    Q3_index = 3 * (n + 1) // 4
    
    # Handle case where there are less than 4 data points
    if n < 4:
        return min(data_copy), max(data_copy)
    
    # Calculate quartiles
    Q1 = data_copy[Q1_index - 1] if (n + 1) % 4 != 0 else (data_copy[Q1_index - 1] + data_copy[Q1_index]) / 2
    Q3 = data_copy[Q3_index - 1] if (n + 1) % 4 != 0 else (data_copy[Q3_index - 1] + data_copy[Q3_index]) / 2
    return Q1, Q3


# Function to calculate IQR without using built-in function
def calculate_iqr(data):
    Q1, Q3 = calculate_quartiles(data)
    return Q3 - Q1

# Function to take user input for number of data points and data
def take_user_input():
    num_data_points = int(input("Enter the number of data points: "))
    data = []
    for i in range(num_data_points):
        num = float(input(f"Enter data point {i + 1}: "))
        data.append(num)
    return data

# Main function
def main():
    data = take_user_input()
    print("Data:", data)
    print("Mean:", calculate_mean(data))
    print("Variance:", calculate_variance(data))
    print("Standard Deviation:", calculate_standard_deviation(data))
    Q1, Q3 = calculate_quartiles(data)
    print("Quartiles:", Q1, Q3)
    print("IQR:", calculate_iqr(data))

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
dataset = pd.read_csv('Mall_Customers.csv')
print(dataset)
 
spending_scores =dataset.iloc[:,3].values
annual_incomes = dataset.iloc[:, 4].values

plt.scatter(spending_scores,annual_incomes)
plt.title('Spending Score vs. Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()

X = dataset.iloc[:, [3, 4]].values
     
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Read data from CSV file
data = pd.read_csv('blobs.csv')
print(data)


# Extract the features (assuming your CSV file has columns 'Feature1' and 'Feature2')
X = data.iloc[:,[0,1]].values
print(X)
     


# DBSCAN clustering
db = DBSCAN(eps=0.5, min_samples=5)
y_db = db.fit_predict(X)
     



# Number of clusters in labels, ignoring noise if present (-1)
n_clusters_ = len(set(y_db)) - (1 if -1 in y_db else 0)
     




# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[y_db == 0][:, 0], X[y_db == 0][:, 1], c='blue', marker='o', label='Cluster 1')
plt.scatter(X[y_db == 1][:, 0], X[y_db == 1][:, 1], c='green', marker='s', label='Cluster 2')
plt.scatter(X[y_db == 2][:, 0], X[y_db == 2][:, 1], c='red', marker='^', label='Cluster 3')
plt.scatter(X[y_db == -1][:, 0], X[y_db == -1][:, 1], c='gray', marker='x', label='Noise')
plt.legend(loc='best')
plt.title(f"DBSCAN Clustering (Number of Clusters: {n_clusters_})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data
data, _ = make_blobs(n_samples=300, centers=4, cluster_std=.60, random_state=0)
     


# Set the grid size (you can adjust this based on your data distribution)
grid_size = 1.0

# Get the minimum and maximum values for x and y coordinates
x_min, x_max = data[:, 0].min(), data[:, 0].max()
y_min, y_max = data[:, 1].min(), data[:, 1].max()

     


# Create a grid by defining intervals using the minimum and maximum values
x_grid = np.arange(x_min, x_max + grid_size, grid_size)
y_grid = np.arange(y_min, y_max + grid_size, grid_size)

     


# Initialize labels array with zeros
labels = np.zeros(data.shape[0], dtype=int)

     


# Assign each data point to a grid cell based on its coordinates
for i, point in enumerate(data):
    x, y = point
    x_label = np.searchsorted(x_grid, x) - 1
    y_label = np.searchsorted(y_grid, y) - 1
    labels[i] = x_label * len(y_grid) + y_label

     


# Visualize the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title('Grid-based Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
