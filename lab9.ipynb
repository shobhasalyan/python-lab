```python
import cv2
import numpy as np

def local_thresholding(image, block_size, C):
    # Create an empty image for the segmented result
    segmented_image = np.zeros_like(image)
    
    # Pad the image to handle edge cases
    padded_image = np.pad(image, ((block_size//2, block_size//2), (block_size//2, block_size//2)), mode='constant')
    
    # Loop through each pixel in the image
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            # Extract the local region around the pixel
            local_region = padded_image[i:i+block_size, j:j+block_size]
            
            # Calculate the local mean and standard deviation
            local_mean = np.mean(local_region)
            local_std = np.std(local_region)
            
            # Calculate the threshold using the formula: T = mean - C * std
            threshold = local_mean - C * local_std
            
            # Apply local thresholding
            if image[i, j] > threshold:
                segmented_image[i, j] = 255
            else:
                segmented_image[i, j] = 0
    
    return segmented_image

# Load the image
image_path=input("Enter image path: ")
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define block size and C value for local thresholding
block_size = int(input("Enter the block size(3-21): "))
C = int(input("Enter the C value(1-20): "))

# Perform local thresholding
segmented_image = local_thresholding(image, block_size, C)

# Display the original and segmented images
cv2.imshow('Original Image', image)
cv2.imshow('Segmented Image (Local Thresholding)', segmented_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def global_thresholding(image, threshold_value):
    _, segmented_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
    return segmented_image

def local_thresholding(image, block_size, C):
    segmented_image = np.zeros_like(image)  # Create an empty image for the segmented result
    padded_image = np.pad(image, block_size // 2, mode='constant')  # Pad the image to handle edge cases
    
    rows, cols = image.shape  # Get dimensions of the original image
    
    for i in range(rows):
        for j in range(cols):
            # Extract the local region around the pixel
            local_region = padded_image[i:i + block_size, j:j + block_size]
            
            # Calculate local mean and standard deviation
            local_mean = np.mean(local_region)
            local_std = np.std(local_region)
            
            # Calculate threshold using the formula: T = mean - C * std
            threshold = local_mean - C * local_std
            
            # Apply local thresholding
            if image[i, j] > threshold:
                segmented_image[i, j] = 255
            else:
                segmented_image[i, j] = 0
                
    return segmented_image

# Load the image
image_path=input("Enter image path: ")
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define parameters for global thresholding
global_threshold_value = 127

# Define parameters for local thresholding
local_block_size = 21
local_C = 0.2

# Perform global thresholding
segmented_global = global_thresholding(image, global_threshold_value)

# Perform local thresholding
segmented_local = local_thresholding(image, local_block_size, local_C)

# Plot the original, globally segmented, and locally segmented images
plt.figure(figsize=(15, 5))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Globally segmented image
plt.subplot(1, 3, 2)
plt.imshow(segmented_global, cmap='gray')
plt.title('Globally Segmented Image')
plt.axis('off')

# Locally segmented image
plt.subplot(1, 3, 3)
plt.imshow(segmented_local, cmap='gray')
plt.title('Locally Segmented Image')
plt.axis('off')


plt.show()
```
